# Visualization of bias and cofidence bands

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
mydf = pd.read_csv('hello.txt', sep=" ", header=None)
mydf.columns = ["true_theta_values", "bias_values", "standard_error_bias_values", "corr_coeff_values"]
```

```python
fig, ax = plt.subplots()
ax.plot(mydf["true_theta_values"],mydf["bias_values"])
ax.fill_between(mydf["true_theta_values"], (mydf["bias_values"]-2*mydf["standard_error_bias_values"]),
                (mydf["bias_values"]+2*mydf["standard_error_bias_values"]), color='b', alpha=.1)
# set x-axis label
ax.set_xlabel(r"True $\theta_0$",fontsize=14)
# set y-axis label
ax.set_ylabel("Estimated bias",color="blue",fontsize=14)


ax2=ax.twinx()
ax2.plot(mydf["true_theta_values"],mydf["corr_coeff_values"], color="red")
# set y-axis label
ax2.set_ylabel(r"Estimated $\rho_{0}$",color="red",fontsize=14)

plt.show()

fig.savefig('IV1.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```

## Visualizing with setting rho to extreme values (0.99 an -0.99)

```python
mydf = pd.read_csv('hello.txt', sep=" ", header=None)
mydf.columns = ["theta", "b_hat_all_ols_mean", "standard_dev1_ols", "bias_corrected_b_hat_lower_ols_mean",
               "standard_dev2_ols", "bias_corrected_b_hat_upper_ols_mean", "standard_dev3_ols", 
               "x_u_corr_ols_mean", "b_hat_all_iv_mean", "standard_dev1_iv", "x_u_corr_iv_mean"]
```

```python
fig, ax = plt.subplots()
ax.plot(mydf["x_u_corr_ols_mean"],mydf["b_hat_all_ols_mean"])
ax.fill_between(mydf["x_u_corr_ols_mean"], (mydf["b_hat_all_ols_mean"]-2*mydf["standard_dev1_ols"]),
                (mydf["b_hat_all_ols_mean"]+2*mydf["standard_dev1_ols"]), color='b', alpha=.1)

ax.plot(mydf["x_u_corr_ols_mean"],mydf["bias_corrected_b_hat_lower_ols_mean"], color="red")
ax.fill_between(mydf["x_u_corr_ols_mean"], (mydf["bias_corrected_b_hat_lower_ols_mean"]-2*mydf["standard_dev2_ols"]),
                (mydf["bias_corrected_b_hat_lower_ols_mean"]+2*mydf["standard_dev2_ols"]), color='r', alpha=.1)

ax.plot(mydf["x_u_corr_ols_mean"],mydf["bias_corrected_b_hat_upper_ols_mean"], color="red")
ax.fill_between(mydf["x_u_corr_ols_mean"], (mydf["bias_corrected_b_hat_upper_ols_mean"]-2*mydf["standard_dev3_ols"]),
                (mydf["bias_corrected_b_hat_upper_ols_mean"]+2*mydf["standard_dev3_ols"]), color='r', alpha=.1)

ax.plot(mydf["x_u_corr_iv_mean"],mydf["b_hat_all_iv_mean"])
ax.fill_between(mydf["x_u_corr_iv_mean"], (mydf["b_hat_all_iv_mean"]-2*mydf["standard_dev1_iv"]),
                (mydf["b_hat_all_iv_mean"]+2*mydf["standard_dev1_iv"]), color='g', alpha=.1)

# set x-axis label
ax.set_xlabel(r"Estimated $\rho_0$",fontsize=14)
# set y-axis label
ax.set_ylabel("Beta esimate",color="blue",fontsize=14)

plt.show()

fig.savefig('IV1.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```

## Visualizing bias as percentages

```python
mydf = pd.read_csv('hello.txt', sep=" ", header=None)
mydf.columns = ["theta", "rho_zero", "bias_as_percentage_all_ols_mean", "standard_dev2_ols",
               "bias_as_percentage_all_iv_mean", "standard_dev2_iv"]
```

```python
fig, ax = plt.subplots()
ax.plot(mydf["rho_zero"],mydf["bias_as_percentage_all_ols_mean"])
ax.fill_between(mydf["rho_zero"], (mydf["bias_as_percentage_all_ols_mean"]-2*mydf["standard_dev2_ols"]),
                (mydf["bias_as_percentage_all_ols_mean"]+2*mydf["standard_dev2_ols"]), color='b', alpha=.1)

ax.plot(mydf["rho_zero"],mydf["bias_as_percentage_all_iv_mean"])
ax.fill_between(mydf["rho_zero"], (mydf["bias_as_percentage_all_iv_mean"]-2*mydf["standard_dev2_iv"]),
                (mydf["bias_as_percentage_all_iv_mean"]+2*mydf["standard_dev2_iv"]), color='orange', alpha=.1)

# set x-axis label
ax.set_xlabel(r"True $\rho_0$" + "\n" + r"True $\theta_0$",fontsize=14)
# set y-axis label
ax.set_ylabel("Estimated bias as % of beta",color="blue",fontsize=14)

fig.canvas.draw()

labels = list(zip(mydf['rho_zero'], mydf['theta']))

final_labels = []
final_ticks = []

counter = 0

for i in range(len(labels)):
    if i == 0:
        final_labels.append(labels[i])
        final_ticks.append(labels[i][0])
    elif labels[i][0] -  final_ticks[counter] > 0.1:
        final_labels.append(labels[i])
        final_ticks.append(labels[i][0])
        counter+=1
        
final_labels = [str(item[0]) + "\n" + str(item[1]) for item in final_labels]
        
ax.set_xticks(final_ticks)

ax.set_xticklabels(final_labels)

plt.show()

fig.savefig('IV1.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
ax.plot(mydf["rho_zero"],mydf["bias_as_percentage_all_ols_mean"])
ax.fill_between(mydf["rho_zero"], (mydf["bias_as_percentage_all_ols_mean"]-2*mydf["standard_dev2_ols"]),
                (mydf["bias_as_percentage_all_ols_mean"]+2*mydf["standard_dev2_ols"]), color='b', alpha=.1)

ax.plot(mydf["rho_zero"],mydf["bias_as_percentage_all_iv_mean"])
ax.fill_between(mydf["rho_zero"], (mydf["bias_as_percentage_all_iv_mean"]-2*mydf["standard_dev2_iv"]),
                (mydf["bias_as_percentage_all_iv_mean"]+2*mydf["standard_dev2_iv"]), color='orange', alpha=.1)

# set x-axis label
ax.set_xlabel(r"True $\rho_0$",fontsize=14)
# set y-axis label
ax.set_ylabel("Estimated bias as % of beta",color="blue",fontsize=14)

ax.set_ylim([-0.4, 0.8])
ax.set_xlim([0, 1])

plt.show()

fig.savefig('IV2.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
