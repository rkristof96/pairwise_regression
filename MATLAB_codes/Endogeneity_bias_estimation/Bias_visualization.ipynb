# Visualization of bias and cofidence bands

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
mydf = pd.read_csv('hello.txt', sep=" ", header=None)
mydf.columns = ["true_theta_values", "bias_values", "standard_error_bias_values", "corr_coeff_values"]
```

```python
fig, ax = plt.subplots()
ax.plot(mydf["true_theta_values"],mydf["bias_values"])
ax.fill_between(mydf["true_theta_values"], (mydf["bias_values"]-2*mydf["standard_error_bias_values"]),
                (mydf["bias_values"]+2*mydf["standard_error_bias_values"]), color='b', alpha=.1)
# set x-axis label
ax.set_xlabel(r"True $\theta_0$",fontsize=14)
# set y-axis label
ax.set_ylabel("Estimated bias",color="blue",fontsize=14)


ax2=ax.twinx()
ax2.plot(mydf["true_theta_values"],mydf["corr_coeff_values"], color="red")
# set y-axis label
ax2.set_ylabel(r"Estimated $\rho_{0}$",color="red",fontsize=14)

plt.show()

fig.savefig('n_0_1.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```

```python
mydf = pd.read_csv('hello.txt', sep=" ", header=None)
mydf.columns = ["true_theta_values", "bias_percentages", "standard_error_bias_percentages", "corr_coeff_values"]
```

```python
fig, ax = plt.subplots()
ax.plot(mydf["corr_coeff_values"],mydf["bias_percentages"])
ax.fill_between(mydf["corr_coeff_values"], (mydf["bias_percentages"]-2*mydf["standard_error_bias_percentages"]),
                (mydf["bias_percentages"]+2*mydf["standard_error_bias_percentages"]), color='b', alpha=.1)
# set x-axis label
ax.set_xlabel(r"Estimated $\rho_{0}$",fontsize=14)
# set y-axis label
ax.set_ylabel(r"Estimated bias as a percentage of estimated $\beta$",color="blue",fontsize=14)

plt.show()

fig.savefig('n_0_100.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
